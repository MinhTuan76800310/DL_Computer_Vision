```python
# ------------------------------------------- PREPARE DATA -----------------------------------------
# ------------------------------------------------------------------------------------
```

```python
def CVS2TrainingData(url):
    """
    Download data and transfer into pandas data frame

    Requirement library:
    import pandas as pd
    Input:
    ----------
    url: Link of data (.csv)

    Output:
    ----------
    X_train:
    Y_train:
    """
    data = pd.read_csv(url)
    data_cols = data.columns # index value
    # print(type(ConcreteDataCol))
    X_train = data[data_cols[data_cols != 'Strength']]
    Y_train = data['Strength']
    return X_train, Y_train
```

```python
url = 'https://cocl.us/concrete_data'
X_train, Y_train = CVS2TrainingData(url)
#X_train.describe()
#Y_train.describe()
```

```python
def Normalize(X_train):
    return (X_train - X_train.mean())/X_train.std()
```

```python
X_nor = Normalize(X_train)
X_nor.head()
```

```python
NO_X_features = X_nor.shape[1]
NO_X_features
```

```python
# ------------------------------------------- BUILD MODEL -----------------------------------------
# ------------------------------------------------------------------------------------
```

```python

def my_model():
    """
    Build a model that satisfies the following requirements:
    - One hidden layer of 10 nodes, and a ReLU activation function

    - Use the adam optimizer and the mean squared error  as the loss function.

    1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the 
    train_test_split
    helper function from Scikit-learn.

    2. Train the model on the training data using 50 epochs.

    3. Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

    4. Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.

    5. Report the mean and the standard deviation of the mean squared errors.
    

    Requirement library:
    from sklearns.metrics import mean_square_error
    from sklearns.model_selections import train_test_split
    from tensorflow.keras.layers import Dense
    Input:
    ----------

    Output:
    ----------
    model
    """
    model = Sequential();
    model.add(Dense(10, activation='relu', input_shape=(NO_X_features,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python

mse_list = []
for i in range(50):
    X_train_, X_test, Y_train_, Y_test = train_test_split(X_nor, Y_train, test_size=0.3, random_state = None)
    model = my_model()
    model.fit(X_train_, Y_train_, epochs = 50, verbose=1)
    Y_pred = model.predict(X_test)
    mse_val = mean_squared_error(Y_test, Y_pred)
    mse_list.append(mse_val)

mean_mse = mean(mse_list)
std_mse = stdev(mse_list)


```

```python
print(f"Mean MSE over 50 iterations: {mean_mse}")
print(f"Standard deviation of MSE: {std_mse}")
```
